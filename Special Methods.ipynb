## Introduction

Special methods (also called **dunder methods** — short for *double underscore*) are predefined methods in Python that allow you to define how objects of your class behave with built-in operations.

They follow the pattern `__methodname__` and are automatically called by Python in response to certain operations.

In this notebook, we will explore the most commonly used special methods with practical examples.

---
## 1. `__init__` — Constructor

The `__init__` method is called automatically when a new object is created. It initialises the object's attributes.

```python
class Student:
    def __init__(self, name, age, gpa):
        self.name = name
        self.age = age
        self.gpa = gpa

# Creating Student objects
s1 = Student("Alice", 20, 3.8)
s2 = Student("Bob", 22, 3.5)

print(f"Student: {s1.name}, Age: {s1.age}, GPA: {s1.gpa}")
print(f"Student: {s2.name}, Age: {s2.age}, GPA: {s2.gpa}")
```

---
## 2. `__str__` and `__repr__` — String Representations

- `__str__`: Returns a **human-readable** string. Called by `print()` and `str()`.
- `__repr__`: Returns an **official/developer** representation. Called by `repr()` and in the REPL.

```python
class Student:
    def __init__(self, name, age, gpa):
        self.name = name
        self.age = age
        self.gpa = gpa

    def __str__(self):
        return f"Student(name={self.name}, age={self.age}, gpa={self.gpa})"

    def __repr__(self):
        return f"Student('{self.name}', {self.age}, {self.gpa})"

s1 = Student("Alice", 20, 3.8)

print(str(s1))   # calls __str__
print(repr(s1))  # calls __repr__
print(s1)        # also calls __str__
```

---
## 3. `__len__` — Length

`__len__` allows the use of the built-in `len()` function on your objects.

```python
class Classroom:
    def __init__(self, students):
        self.students = students  # a list of student names

    def __len__(self):
        return len(self.students)

    def __str__(self):
        return f"Classroom with {len(self)} students"

room = Classroom(["Alice", "Bob", "Charlie", "Diana"])
print(len(room))  # calls __len__
print(room)
```

## 4. Comparison Methods — `__eq__`, `__lt__`, `__gt__`, `__le__`, `__ge__`

These methods define how objects are compared using `==`, `<`, `>`, `<=`, `>=`.

```python
class Student:
    def __init__(self, name, gpa):
        self.name = name
        self.gpa = gpa

    def __eq__(self, other):
        return self.gpa == other.gpa

    def __lt__(self, other):
        return self.gpa < other.gpa

    def __gt__(self, other):
        return self.gpa > other.gpa

    def __le__(self, other):
        return self.gpa <= other.gpa

    def __ge__(self, other):
        return self.gpa >= other.gpa

    def __str__(self):
        return f"{self.name} (GPA: {self.gpa})"

s1 = Student("Alice", 3.8)
s2 = Student("Bob", 3.5)
s3 = Student("Carol", 3.8)

print(f"{s1} == {s3}: {s1 == s3}")  # True — same GPA
print(f"{s1} == {s2}: {s1 == s2}")  # False
print(f"{s2} < {s1}: {s2 < s1}")    # True
print(f"{s1} > {s2}: {s1 > s2}")    # True

# We can even sort a list of students by GPA!
students = [s1, s2, s3]
students.sort()
print("\nSorted by GPA:")
for s in students:
    print(" ", s)
```

## 5. Arithmetic Methods — `__add__`, `__sub__`, `__mul__`

These let you use `+`, `-`, `*` operators on your custom objects.

```python
class Vector:
    """Represents a 2D mathematical vector."""

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def __mul__(self, scalar):
        return Vector(self.x * scalar, self.y * scalar)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(1, 5)

print(f"v1 = {v1}")
print(f"v2 = {v2}")
print(f"v1 + v2 = {v1 + v2}")   # calls __add__
print(f"v1 - v2 = {v1 - v2}")   # calls __sub__
print(f"v1 * 3  = {v1 * 3}")    # calls __mul__
```

## 6. `__getitem__`, `__setitem__`, `__delitem__` — Index Access

These methods allow objects to support indexing with `[]`.

```python
class GradeBook:
    def __init__(self):
        self.grades = {}

    def __setitem__(self, student, grade):
        self.grades[student] = grade

    def __getitem__(self, student):
        return self.grades.get(student, "Not found")

    def __delitem__(self, student):
        if student in self.grades:
            del self.grades[student]
            print(f"{student}'s grade deleted.")
        else:
            print(f"{student} not in grade book.")

    def __str__(self):
        return str(self.grades)

gb = GradeBook()
gb["Alice"] = 95       # calls __setitem__
gb["Bob"] = 87
gb["Charlie"] = 91

print(gb["Alice"])     # calls __getitem__
print(gb["Diana"])     # returns 'Not found'
print(gb)

del gb["Bob"]          # calls __delitem__
print(gb)
```

## 7. `__contains__` — Membership Test (`in` operator)

Allows the use of `in` and `not in` operators on your objects.

```python
class Classroom:
    def __init__(self, students):
        self.students = students

    def __contains__(self, name):
        return name in self.students

    def __len__(self):
        return len(self.students)

    def __str__(self):
        return f"Classroom: {self.students}"

room = Classroom(["Alice", "Bob", "Charlie"])

print("Alice" in room)    # calls __contains__ → True
print("Diana" in room)    # False
print("Bob" not in room)  # False
```

## 8. `__iter__` and `__next__` — Iteration

These methods make your objects **iterable** (usable in `for` loops).

```python
class CountDown:
    """Counts down from a given number to 1."""

    def __init__(self, start):
        self.start = start
        self.current = start

    def __iter__(self):
        self.current = self.start  # reset on each iteration
        return self

    def __next__(self):
        if self.current <= 0:
            raise StopIteration
        value = self.current
        self.current -= 1
        return value

cd = CountDown(5)

print("Counting down:")
for num in cd:
    print(num)

# We can iterate again
print("\nAgain:", list(cd))
```

## 9. `__call__` — Making Objects Callable

`__call__` lets you call an object like a function.

```python
class Multiplier:
    """An object that multiplies numbers by a fixed factor."""

    def __init__(self, factor):
        self.factor = factor

    def __call__(self, value):
        return value * self.factor

double = Multiplier(2)
triple = Multiplier(3)

print(double(5))   # calls __call__ → 10
print(triple(5))   # calls __call__ → 15
print(double(10))  # 20

# Check if an object is callable
print(callable(double))  # True
```

## 10. `__enter__` and `__exit__` — Context Manager (`with` statement)

These methods allow your objects to be used with the `with` statement for resource management.

```python
class DatabaseConnection:
    """Simulates a database connection using a context manager."""

    def __init__(self, db_name):
        self.db_name = db_name

    def __enter__(self):
        print(f"[+] Connecting to database: {self.db_name}")
        return self  # returned as 'as' variable

    def __exit__(self, exc_type, exc_value, traceback):
        print(f"[-] Closing connection to: {self.db_name}")
        return False  # don't suppress exceptions

    def query(self, sql):
        print(f"  Executing: {sql}")

# The connection opens and closes automatically
with DatabaseConnection("students.db") as db:
    db.query("SELECT * FROM students")
    db.query("SELECT * FROM courses")

print("Done — connection was closed automatically.")
```

## 11. Putting It All Together — A Full Example

Below is a `BankAccount` class that uses several special methods together.

```python
class BankAccount:
    """A simple bank account with special method support."""

    def __init__(self, owner, balance=0.0):
        self.owner = owner
        self.balance = balance
        self.transactions = []

    def __str__(self):
        return f"BankAccount(owner={self.owner}, balance=${self.balance:.2f})"

    def __repr__(self):
        return f"BankAccount('{self.owner}', {self.balance})"

    def __len__(self):
        return len(self.transactions)

    def __add__(self, amount):
        """Deposit: account + amount"""
        new_acc = BankAccount(self.owner, self.balance + amount)
        new_acc.transactions = self.transactions + [f"Deposit: +${amount}"]
        return new_acc

    def __sub__(self, amount):
        """Withdraw: account - amount"""
        if amount > self.balance:
            raise ValueError("Insufficient funds")
        new_acc = BankAccount(self.owner, self.balance - amount)
        new_acc.transactions = self.transactions + [f"Withdrawal: -${amount}"]
        return new_acc

    def __eq__(self, other):
        return self.balance == other.balance

    def __lt__(self, other):
        return self.balance < other.balance

    def __gt__(self, other):
        return self.balance > other.balance

    def __contains__(self, keyword):
        """Check if a keyword appears in any transaction."""
        return any(keyword in t for t in self.transactions)

    def __iter__(self):
        return iter(self.transactions)


# Demo
acc = BankAccount("Alice", 1000.0)
print(acc)

acc = acc + 500   # deposit
acc = acc + 200
acc = acc - 300   # withdrawal

print(acc)
print(f"Number of transactions: {len(acc)}")

print("\nTransaction History:")
for t in acc:     # uses __iter__
    print(" ", t)

print("\n'Deposit' in transactions:", "Deposit" in acc)  # __contains__

acc2 = BankAccount("Bob", 800.0)
print(f"\n{acc.owner} richer than {acc2.owner}: {acc > acc2}")  # __gt__
```
